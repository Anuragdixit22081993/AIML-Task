```python
import pandas as pd
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.utils import plot_model
```

```python
# Load training data
train_url = "https://raw.githubusercontent.com/Anuragdixit22081993/my_practice_aiml/refs/heads/main/guru_assignments/dataset/datasets/pima-indians_classification_train.csv"
train_df = pd.read_csv(train_url)
```

```python
# Split features and target
X_train = train_df.iloc[:, :-1]
y_train = train_df.iloc[:, -1]
```

```python
# Build Neural Network
model = Sequential()
```

```python

# Input layer + Hidden Layer 1
model.add(Dense(
    12,
    input_dim=8,
    activation='relu',
    name='Hidden_Layer_1'
))
```

```python
# Hidden Layer 2
model.add(Dense(
    8,
    activation='relu',
    name='Hidden_Layer_2'
))
```

```python
# Output Layer
model.add(Dense(
    1,
    activation='sigmoid',
    name='Output_Layer'
))
```

```python
# Compile Model
model.compile(
    loss='binary_crossentropy',
    optimizer='adam',
    metrics=['accuracy']
)
```

```python
# Print Architecture
model.summary()

# Draw Network Architecture
plot_model(
    model,
    to_file='pima_neural_network.png',
    show_shapes=True,
    show_layer_names=True,
    rankdir='LR'
)

print("Neural network diagram saved as pima_neural_network.png")
```
